# Random Graph
## Small World (Watts-Strogatz)

```python
fig = plt.figure(figsize=(15, 15))
plt.subplot(221, title='k=4 p=0')
graph1 = nx.watts_strogatz_graph(n = 14, k = 4, p = 0, seed = 3)
nx.draw_circular(graph1)

plt.subplot(222, title='k=4 p=0.3')
graph2 = nx.watts_strogatz_graph(n = 14, k = 4, p = 0.3)
nx.draw_circular(graph2)

plt.subplot(223, title='k=2 p=0')
graph3 = nx.watts_strogatz_graph(n = 14, k = 2, p = 0)
nx.draw_circular(graph3)

plt.subplot(224, title='k=2 p=0.3')
graph4 = nx.watts_strogatz_graph(n = 14, k = 2, p = 0.3)
nx.draw_circular(graph4)
```

## Small World (Newman-Watts-Strogatz)

```python
fig = plt.figure(figsize=(15, 15))
plt.subplot(221, title='k=4 p=0')
graph1 = nx.newman_watts_strogatz_graph(n = 14, k = 4, p = 0, seed = 3)
nx.draw_circular(graph1)

plt.subplot(222, title='k=4 p=0.3')
graph2 = nx.newman_watts_strogatz_graph(n = 14, k = 4, p = 0.3)
nx.draw_circular(graph2)

plt.subplot(223, title='k=2 p=0')
graph3 = nx.newman_watts_strogatz_graph(n = 14, k = 2, p = 0)
nx.draw_circular(graph3)

plt.subplot(224, title='k=2 p=0.3')
graph4 = nx.newman_watts_strogatz_graph(n = 14, k = 2, p = 0.3)
nx.draw_circular(graph4)
```

## Barabasi-Albert graph

## Erdos-Renyi

```python
graph = nx.erdos_renyi_graph(n = 10, p = 0.3, seed=4, directed=False)
nx.draw(graph)
```

# Bipartite

```python
from networkx.algorithms import bipartite
graph = bipartite.random_graph(n = 8, m = 6, p = 0.5, seed = 3, directed = False)

top = nx.bipartite.sets(graph)[0]
pos = nx.bipartite_layout(graph, top)
nx.draw(graph, pos)
```

## Scale Free

```python
graph = nx.scale_free_graph(20, alpha=0.41, beta=0.54, gamma=0.05, delta_in=0.2, delta_out=0, create_using=None, seed=4)
nx.draw(graph)
```
